```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
# import data as dataframe
wind_citypop_df = pd.read_csv("Resources/wind_citypop.csv") 
wind_citypop_df = wind_citypop_df.drop(['Unnamed: 0'],axis=1) 
wind_citypop_df.head()
```

```python
wind_citypop_df.info()
```

```python
# We investigate the datas which population Density no more than 500 per square mile
wind_citypop_df= wind_citypop_df[wind_citypop_df['Density']<=500.0]
wind_state_df = wind_citypop_df.groupby('State')['City/Town/County'].count()
wind_state_df = wind_state_df.reset_index() 
wind_state_df = wind_state_df.sort_values(by="City/Town/County", ascending=False )
wind_state_df = wind_state_df.rename(columns={"City/Town/County": "Suitable Sites"})
wind_state_df.head()
```

```python
# create x_axis and ticks locations
x_axis=np.arange(len(wind_state_df))
tick_locations = [i for i in x_axis ]
# Create a bar chart based on wind_state_df
plt.figure(figsize=(10,5))
plt.bar(x_axis,wind_state_df['Suitable Sites'],color='g', alpha=0.5, align="center",width=0.7)
plt.xticks(tick_locations, wind_state_df["State"], rotation="vertical")

# Set x and y limits
plt.xlim(-0.75, len(x_axis))
plt.ylim(0, max(wind_state_df['Suitable Sites'])+4)

# Set title , xlabel, ylabel
plt.xlabel('States')
plt.ylabel('Suitable Sites')
plt.title("Suitable Sites in Each States")

# Set the grid
plt.grid()

# Save our graph and show the grap
plt.tight_layout()
plt.savefig("Image/Suitable Sites in Each States.png")
plt.show()
```

```python
#import other csv file as dataframe
# Energy Census and Economic Data US 2010-2014
Energy_Census_df = pd.read_csv('Resources/Energy Census and Economic Data US 2010-2014.csv')
Energy_Census_df.head()
```

```python
# We only retrieve "StateCodes" and 'CoalC2014' from the original dataset
Energy_Census_df = Energy_Census_df[['StateCodes','CoalC2014']]
```

```python
#Delete unwanted rows, sort the values for CoalC2014 column and rename column StateCodes
Energy_Census_df = Energy_Census_df[Energy_Census_df['StateCodes']!='US']
Energy_Census_df = Energy_Census_df.sort_values(by = 'CoalC2014',ascending=False)
Energy_Census_df = Energy_Census_df.rename(columns= {"StateCodes":"State"})
Energy_Census_df.head(10)
```

```python
# Source: https://www.eia.gov/coal/production/quarterly/co2_article/co2.html
# "CoalC2014": Coal total consumption in billion BTU in 2014
# Btu :British thermal units
# Coal emits about 204.3 pounds of carbon dioxide per million Btu when completely burned.
# 2000 lb/ 1 tons

#Here we investigate carbon dioxide production by coal For each state in 2014
CO2_2014produc_df = Energy_Census_df
# Covert Coal consumption to Carbon Dioxide production (ton)
CO2_2014produc_df['CoalC2014'] = (CO2_2014produc_df['CoalC2014']*204300)/2000 
CO2_2014produc_df['CoalC2014']= round(CO2_2014produc_df['CoalC2014'], 0)
CO2_2014produc_df = CO2_2014produc_df.rename(columns={'CoalC2014':'CO2 Production 2014(ton)'})
CO2_2014produc_df.head(10)
```

```python
# Plot carbon dioxide production by coal for each state
data = dict(type = 'choropleth',
            colorscale = 'Bluered',
            locations = CO2_2014produc_df['State'],
            locationmode = 'USA-states',
            z = CO2_2014produc_df['CO2 Production 2014(ton)'],
            text = CO2_2014produc_df['State'],
            marker = dict(line = dict(color = 'rgb(255,255,255)', width = 2)),
            colorbar = {'title': 'Tons of C02'})
layout = dict(title = 'Carbon Dioxide Production by Coal in 2014',
             geo = dict(scope = 'usa',showlakes = True, lakecolor = 'rgb(85, 173, 240)'))
choromap = go.Figure(data = [data], layout=layout)
iplot(choromap)
```

```python
# Combine two dataframes 
CO2_state_df = CO2_2014produc_df.merge(wind_state_df,on='State',how='left')
# Replace nan value with 0
CO2_state_df = CO2_state_df.fillna(0)
# change the value type
CO2_state_df['Suitable Sites']= CO2_state_df['Suitable Sites'].astype(int)
#Sort the column for CO2 Production 2014
CO2_state_df= CO2_state_df.sort_values(by='CO2 Production 2014(ton)',ascending=False)
CO2_state_df = CO2_state_df.reset_index()[['State','CO2 Production 2014(ton)','Suitable Sites']]
CO2_state_df.head(10)
```

```python
# Plot the CO2 Production 
# create x_axis and ticks locations
x_axis=np.arange(len(CO2_state_df))
tick_locations = [i for i in x_axis ]

# Create a bar chart based on wind_state_df_2
plt.figure(figsize=(14,5))
plt.bar(x_axis,CO2_state_df['CO2 Production 2014(ton)'],color='red', alpha=0.6, align="center")
plt.xticks(tick_locations, CO2_state_df["State"], rotation="vertical")

# Set x and y limits
plt.xlim(-0.75, len(x_axis))


# Set title , xlabel
plt.xlabel('States')
plt.ylabel("CO2 Production (ton)")
plt.title("CO2 Production")

# Create Text 
# Set The position to place the text and the text we need to add
text_xloc=[x for x in x_axis]
text_yloc=[x for x in CO2_state_df['CO2 Production 2014(ton)']]
text=[x for x in CO2_state_df['Suitable Sites']]
for i in range(len(CO2_state_df)):
    plt.text(text_xloc[i],text_yloc[i],text[i],ha="center", va="bottom", color = "green")



# Save our graph and show the grap
plt.tight_layout()
plt.savefig("Image/CO2 Production.png")

```
